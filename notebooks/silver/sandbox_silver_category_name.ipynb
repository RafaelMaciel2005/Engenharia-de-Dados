# Sandbox — Silver Category Name

Exploração por trás de `silver_rules.tratar_category_name`. É a menor tabela do dataset: uma **tabela de tradução** que mapeia a categoria em português para o inglês. O tratamento é simples — normalizar as duas colunas em minúsculo/trim para garantir joins consistentes com `products` (cuja `product_category_name` também é normalizada na Silver).

```python
from utils.spark_utils import create_spark_session
from pyspark.sql.functions import col, trim, lower, current_timestamp

spark = create_spark_session('Sandbox-Silver-Category-Name')

df_bronze = spark.read.parquet('s3a://bronze/olist/category_name')
df_bronze.show(10, truncate=False)
df_bronze.printSchema()
```

Esta tabela é pequena o suficiente para conferir o total de categorias e garantir que não há duplicidade na chave de tradução.

```python
print('Total de categorias:', df_bronze.count())
print('Categorias distintas:', df_bronze.select('product_category_name').distinct().count())
```

```python
df_silver = (df_bronze
    .withColumn('product_category_name', lower(trim(col('product_category_name'))))
    .withColumn('product_category_name_english', lower(trim(col('product_category_name_english'))))
    .withColumn('dt_criacao_silver', current_timestamp()))

df_silver.printSchema()
df_silver.show(10, truncate=False)
```
